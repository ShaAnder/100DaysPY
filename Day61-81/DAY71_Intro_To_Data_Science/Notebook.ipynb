```python
import pandas as pd
```

```python
df = pd.read_csv('salaries_by_college_major.csv')
pd.options.display.float_format = '{:,.2f}'.format 
```

```python

#identify bad cells
print(df.isna())

#clean the df
clean_df = df.dropna()
print(clean_df.tail())
print(clean_df)
```

```python

#check df columns
print(clean_df['Starting Median Salary'])
```

```python
#can dive deeper into this with aditional arguments
print(clean_df['Starting Median Salary'].idxmax(), clean_df['Starting Median Salary'].max())
```

```python

#unless specifying the column we can just get the entire entry with
print(clean_df.loc[43])
```

```python
### --- CHALLENGE 1 --- ###

# get highest mid career salary, find how much a graduate of this major earns
x = clean_df['Mid-Career Median Salary'].idxmax()
print(  f"Row {x} \n \
        Name: {clean_df.loc[x]['Undergraduate Major']}, \n \
        The Mid-Career Salary is: {clean_df.loc[x]['Mid-Career Median Salary']}, \n \
        The graduate starting salary is: {clean_df.loc[x]['Starting Median Salary']}"
        )

#find lowest mid career salary find out the potential people can expect to earn in this field
y = clean_df['Mid-Career Median Salary'].idxmin()
print(  f"Row {y} \n \
        Name: {clean_df.loc[y]['Undergraduate Major']}, \n \
        The Mid-Career Salary is: {clean_df.loc[y]['Mid-Career Median Salary']}, \n \
        The Mid Career High End salary is: {clean_df.loc[y]['Mid-Career 90th Percentile Salary']}"
        )

#find the lowest starting salary and find the expected maximum
z = clean_df['Starting Median Salary'].idxmin()
print(  f"Row {z} \n \
        Name: {clean_df.loc[z]['Undergraduate Major']}, \n \
        The Starting Salary is: {clean_df.loc[z]['Mid-Career Median Salary']}, \n \
        The Mid Career High End salary is: {clean_df.loc[z]['Mid-Career 90th Percentile Salary']}"
        )
```

```python
# we can also go and find the differences between columns and then add that as a new column to the df
# eg let's find the potential difference between mid and max salaries
spread = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
#we can then add this subtraction as a new column
clean_df.insert(1, 'Spread', spread)
print(clean_df.head())
#we can then ofc sort by lowest spread
low_risk = clean_df.sort_values('Spread')
print(low_risk[['Undergraduate Major', 'Spread']].head())
```

```python
### --- CHALLENGE 2 --- ###

#find top 5 degrees with highest 90% value
high_value = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
print(high_value[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head())

#find degrees with greatest spread in salary between high and low earners
high_risk = clean_df.sort_values('Spread', ascending=False)
print(high_risk[['Undergraduate Major', 'Spread']].head())
```

```python
# grouping data by category
# we want to count the groups 
print(clean_df.groupby('Group').count())
# mini challenge find the mean salary by group (we stipulate to do this to numeric only so no errors are thrown)
print(clean_df.groupby('Group').mean(numeric_only = True))
```
